### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Calcular o faturamento total para cada loja
faturamento_loja1 = loja['Preço'].sum()
faturamento_loja2 = loja2['Preço'].sum()
faturamento_loja3 = loja3['Preço'].sum()
faturamento_loja4 = loja4['Preço'].sum()

# Imprimir o faturamento total para cada loja
print(f"Faturamento Total da Loja 1: R${faturamento_loja1:,.2f}")
print(f"Faturamento Total da Loja 2: R${faturamento_loja2:,.2f}")
print(f"Faturamento Total da Loja 3: R${faturamento_loja3:,.2f}")
print(f"Faturamento Total da Loja 4: R${faturamento_loja4:,.2f}")

# Preparar dados para o gráfico de faturamento
store_names = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
revenues = [faturamento_loja1, faturamento_loja2, faturamento_loja3, faturamento_loja4]

# Criar um gráfico de barras para o Faturamento Total
plt.figure(figsize=(10, 6))
plt.bar(store_names, revenues, color='skyblue')
plt.title('Faturamento Total por Loja')
plt.xlabel('Loja')
plt.ylabel('Faturamento (R$)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.ticklabel_format(style='plain', axis='y') # Desativar notação científica no eixo y
plt.tight_layout()
plt.savefig('faturamento_total_por_loja.png') # Salvar o gráfico
plt.show() # Mostrar o gráfico (em ambientes como Colab ou Jupyter)
```

# 2. Vendas por Categoria


```python
# Contar as vendas por categoria para cada loja
vendas_por_categoria_loja1 = loja1['Categoria do Produto'].value_counts()
vendas_por_categoria_loja2 = loja2['Categoria do Produto'].value_counts()
vendas_por_categoria_loja3 = loja3['Categoria do Produto'].value_counts()
vendas_por_categoria_loja4 = loja4['Categoria do Produto'].value_counts()

# Plotar as vendas por categoria para a Loja 1
plt.figure(figsize=(12, 7))
vendas_por_categoria_loja1.head(10).sort_values(ascending=True).plot(kind='barh', color='darkblue')
plt.title('Top 10 Vendas por Categoria - Loja 1')
plt.xlabel('Número de Vendas')
plt.ylabel('Categoria do Produto')
plt.tight_layout()
plt.savefig('vendas_por_categoria_loja1.png')

# Plotar as vendas por categoria para a Loja 2
plt.figure(figsize=(12, 7))
vendas_por_categoria_loja2.head(10).sort_values(ascending=True).plot(kind='barh', color='darkgreen')
plt.title('Top 10 Vendas por Categoria - Loja 2')
plt.xlabel('Número de Vendas')
plt.ylabel('Categoria do Produto')
plt.tight_layout()
plt.savefig('vendas_por_categoria_loja2.png')

# Plotar as vendas por categoria para a Loja 3
plt.figure(figsize=(12, 7))
vendas_por_categoria_loja3.head(10).sort_values(ascending=True).plot(kind='barh', color='darkred')
plt.title('Top 10 Vendas por Categoria - Loja 3')
plt.xlabel('Número de Vendas')
plt.ylabel('Categoria do Produto')
plt.tight_layout()
plt.savefig('vendas_por_categoria_loja3.png')

# Plotar as vendas por categoria para a Loja 4
plt.figure(figsize=(12, 7))
vendas_por_categoria_loja4.head(10).sort_values(ascending=True).plot(kind='barh', color='darkorange')
plt.title('Top 10 Vendas por Categoria - Loja 4')
plt.xlabel('Número de Vendas')
plt.ylabel('Categoria do Produto')
plt.tight_layout()
plt.savefig('vendas_por_categoria_loja4.png')

print("Gráficos de 'Vendas por Categoria' para cada loja foram gerados e salvos como PNG.")
```

# 3. Média de Avaliação das Lojas

```python
# Calcular a média de avaliação da compra para cada loja
media_avaliacao_loja1 = loja1['Avaliação da compra'].mean()
media_avaliacao_loja2 = loja2['Avaliação da compra'].mean()
media_avaliacao_loja3 = loja3['Avaliação da compra'].mean()
media_avaliacao_loja4 = loja4['Avaliação da compra'].mean()

# Imprimir as médias de avaliação (Loja 1 será apenas "Loja")
print(f"\nMédia de Avaliação da Loja: {media_avaliacao_loja1:.2f}")
print(f"Média de Avaliação da Loja 2: {media_avaliacao_loja2:.2f}")
print(f"Média de Avaliação da Loja 3: {media_avaliacao_loja3:.2f}")
print(f"Média de Avaliação da Loja 4: {media_avaliacao_loja4:.2f}")

# Preparar dados para o gráfico de média de avaliação (Loja 1 será apenas "Loja")
store_names = ['Loja', 'Loja 2', 'Loja 3', 'Loja 4']
avg_ratings = [media_avaliacao_loja1, media_avaliacao_loja2, media_avaliacao_loja3, media_avaliacao_loja4]

# Criar um gráfico de barras para a Média de Avaliação dos Clientes
plt.figure(figsize=(10, 6))
plt.bar(store_names, avg_ratings, color='lightgreen')
plt.title('Média de Avaliação dos Clientes por Loja')
plt.xlabel('Loja')
plt.ylabel('Avaliação Média (1-5)')
plt.ylim(0, 5) # Definir limite para o eixo y (avaliações são de 1 a 5)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('media_avaliacao_por_loja.png') # Salvar o gráfico
plt.show() # Mostrar o gráfico (em ambientes como Colab ou Jupyter)
```

# 4. Produtos Mais e Menos Vendidos

```python
# Lista de DataFrames e nomes de lojas para iterar e ajustar o nome da Loja 1
lojas = [loja1, loja2, loja3, loja4]
nomes_lojas = ['Loja', 'Loja 2', 'Loja 3', 'Loja 4']

print("--- Análise de Produtos Mais e Menos Vendidos ---")

for i, loja_df in enumerate(lojas):
    nome_loja = nomes_lojas[i]

    # Contar as vendas de cada produto
    vendas_por_produto = loja_df['Produto'].value_counts()

    # Produtos Mais Vendidos (Top 10)
    top_10_produtos = vendas_por_produto.head(10)
    print(f"\nTop 10 Produtos Mais Vendidos - {nome_loja}:")
    print(top_10_produtos)

    # Produtos Menos Vendidos (Bottom 10)
    bottom_10_produtos = vendas_por_produto.tail(10)
    print(f"\nTop 10 Produtos Menos Vendidos - {nome_loja}:")
    print(bottom_10_produtos)

    # Plotar Top 10 Produtos Mais Vendidos
    plt.figure(figsize=(12, 7))
    top_10_produtos.sort_values(ascending=True).plot(kind='barh', color='darkblue')
    plt.title(f'Top 10 Produtos Mais Vendidos - {nome_loja}')
    plt.xlabel('Número de Vendas')
    plt.ylabel('Produto')
    plt.tight_layout()
    plt.savefig(f'top10_produtos_mais_vendidos_{nome_loja.lower().replace(" ", "_")}.png')

    # Plotar Top 10 Produtos Menos Vendidos
    plt.figure(figsize=(12, 7))
    # Para "menos vendidos", podemos querer ordenar do menor para o maior para que o menos vendido apareça no topo do gráfico
    bottom_10_produtos.sort_values(ascending=False).plot(kind='barh', color='darkred')
    plt.title(f'Top 10 Produtos Menos Vendidos - {nome_loja}')
    plt.xlabel('Número de Vendas')
    plt.ylabel('Produto')
    plt.tight_layout()
    plt.savefig(f'top10_produtos_menos_vendidos_{nome_loja.lower().replace(" ", "_")}.png')

print("\nGráficos de 'Produtos Mais e Menos Vendidos' para cada loja foram gerados e salvos como PNG.")
```

# 5. Frete Médio por Loja


```python
# --- Análise e Visualização do Frete Médio por Loja ---

# Calcular a média de frete para cada loja
media_frete_loja1 = loja1['Frete'].mean()
media_frete_loja2 = loja2['Frete'].mean()
media_frete_loja3 = loja3['Frete'].mean()
media_frete_loja4 = loja4['Frete'].mean()

# Imprimir as médias de frete (Loja 1 será apenas "Loja")
print(f"\nMédia de Frete da Loja: R${media_frete_loja1:,.2f}")
print(f"Média de Frete da Loja 2: R${media_frete_loja2:,.2f}")
print(f"Média de Frete da Loja 3: R${media_frete_loja3:,.2f}")
print(f"Média de Frete da Loja 4: R${media_frete_loja4:,.2f}")

# Preparar dados para o gráfico de frete médio (Loja 1 será apenas "Loja")
store_names = ['Loja', 'Loja 2', 'Loja 3', 'Loja 4']
avg_shipping_costs = [media_frete_loja1, media_frete_loja2, media_frete_loja3, media_frete_loja4]

# Criar um gráfico de barras para a Média de Frete
plt.figure(figsize=(10, 6))
plt.bar(store_names, avg_shipping_costs, color='salmon')
plt.title('Média de Frete por Loja')
plt.xlabel('Loja')
plt.ylabel('Frete Médio (R$)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.ticklabel_format(style='plain', axis='y') # Desativar notação científica no eixo y
plt.tight_layout()
plt.savefig('media_frete_por_loja.png') # Salvar o gráfico
plt.show() # Mostrar o gráfico (em ambientes como Colab ou Jupyter)
```
